```python
df=dt.set_index("Date")
```

```python
airpods_data = df[df['Apple_products'] == 'Airpods']
```

```python
airpods_data.head()
```

```python
alpha = 0.5
beta = 0.3


airpods_data['DoubleExpo1'] = 0.0


airpods_data['DoubleExpo1'][0] = airpods_data['Sales'][0]  # Initialization

for i in range(1, len(airpods_data)):
    airpods_data['DoubleExpo1'][i] = alpha * airpods_data['Sales'][i] + \
                                        (1 - alpha) * (airpods_data['DoubleExpo1'][i - 1] +
                                                       beta * (airpods_data['Sales'][i] - airpods_data['DoubleExpo1'][i - 1]))
```

```python
airpods_data.head()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
import numpy as np
```

```python
def evaluate_forecast(actual_values, forecast_values):

    RSFE = np.sum(actual_values - forecast_values)
    MAE = mean_absolute_error(actual_values, forecast_values)
    MSE = mean_squared_error(actual_values, forecast_values)
    RMSE = np.sqrt(MSE)
    mask = actual_values != 0
    MAPE = np.mean(np.abs((actual_values[mask] - forecast_values[mask]) / actual_values[mask])) * 100
    MPE = np.mean((actual_values[mask] - forecast_values[mask]) / actual_values[mask]) * 100

    evaluation_metrics = {
        'RSFE': RSFE,
        'MAE': MAE,
        'MSE': MSE,
        'RMSE': RMSE,
        'MAPE': MAPE,
        'MPE': MPE
    }

    return evaluation_metrics
```

```python
DoubleExpo1_AP_evaluation =evaluate_forecast(airpods_data['Sales'],airpods_data['DoubleExpo1'])
print(DoubleExpo1_AP_evaluation)
```

```python
alpha = 0.3
beta = 0.6
gamma = 0.45


seasonal_period = 12


airpods_data['TripleExpo1'] = 0.0


level = airpods_data['Sales'][0]
trend = 0
seasonal = [airpods_data['Sales'][i] for i in range(seasonal_period)]
seasonal_avg = sum(seasonal) / float(seasonal_period)

for i in range(len(airpods_data)):
    value = airpods_data['Sales'][i]


    last_level = level
    level = alpha * (value - seasonal[i % seasonal_period]) + (1 - alpha) * (level + trend)


    trend = beta * (level - last_level) + (1 - beta) * trend


    seasonal[i % seasonal_period] = gamma * (value - level) + (1 - gamma) * seasonal[i % seasonal_period]


    airpods_data['TripleExpo1'][i] = level + trend + seasonal[i % seasonal_period]
```

```python
TripeExpo1_Ap_evaluation =evaluate_forecast(airpods_data['Sales'],airpods_data['TripleExpo1'])
print(TripeExpo1_Ap_evaluation)
```
